```python
# Cell 4: IBM Watsonx AI Simulation
# ================================
# 3. IBM WATSONX AI SIMULATION
# ================================

from typing import List, Dict  # ✅ Required for type hints

class IBMGraniteAI:
    """
    Simulated IBM Granite foundation model for natural language understanding
    In production, this would connect to actual Watsonx AI Studio
    """

    def __init__(self):
        self.model_name = "granite-13b-chat-v2"
        self.capabilities = [
            "Natural Language Understanding",
            "Intent Classification",
            "Skill Assessment",
            "Personalized Recommendations"
        ]

    def understand_user_intent(self, user_input: str) -> Dict:
        """Analyze user input to extract learning intent"""
        # Simulated NLP processing
        intents = {
            'learn': ['learn', 'study', 'master', 'understand'],
            'career': ['career', 'job', 'professional', 'work'],
            'skill': ['skill', 'ability', 'knowledge', 'expertise'],
            'beginner': ['beginner', 'new', 'start', 'basic'],
            'intermediate': ['intermediate', 'some experience', 'familiar'],
            'advanced': ['advanced', 'expert', 'experienced', 'proficient']
        }

        detected_intent = {}
        user_lower = user_input.lower()

        for intent, keywords in intents.items():
            if any(keyword in user_lower for keyword in keywords):
                detected_intent[intent] = True

        return detected_intent

    def assess_skill_level(self, user_responses: List[str]) -> str:
        """Assess user's current skill level based on responses"""
        # Simulated skill assessment
        if any('never' in response.lower() or 'no experience' in response.lower()
               for response in user_responses):
            return 'Beginner'
        elif any('some' in response.lower() or 'few projects' in response.lower()
                 for response in user_responses):
            return 'Intermediate'
        else:
            return 'Advanced'

    def generate_response(self, prompt: str, context: Dict = None) -> str:
        """Generate contextual AI responses using Granite model"""
        # Simulated AI response generation
        responses = {
            'greeting': "Hello! I'm LearnMate, powered by IBM Granite AI. I'm here to help you discover the perfect learning path for your goals! 🎯",
            'interest_inquiry': "What domain are you most excited to explore? I can help with Frontend Development, Cybersecurity, UI/UX Design, Data Science, and many more!",
            'skill_assessment': "Let me understand your current level. Could you tell me about any previous experience or projects you've worked on?",
            'recommendation': "Based on our conversation, I've crafted a personalized learning roadmap just for you! Let me show you the exciting journey ahead.",
            'encouragement': "You're making excellent progress! Remember, every expert was once a beginner. Keep up the great work! 💪"
        }

        if 'greeting' in prompt.lower():
            return responses['greeting']
        elif 'interest' in prompt.lower() or 'domain' in prompt.lower():
            return responses['interest_inquiry']
        elif 'skill' in prompt.lower() or 'level' in prompt.lower():
            return responses['skill_assessment']
        elif 'recommend' in prompt.lower():
            return responses['recommendation']
        else:
            return responses['encouragement']

# Initialize IBM Granite AI
granite_ai = IBMGraniteAI()
print(f"\n🤖 IBM Granite AI Model Initialized: {granite_ai.model_name}")
print(f"Capabilities: {', '.join(granite_ai.capabilities)}")
```

```python
from dataclasses import dataclass
from typing import List, Dict
from datetime import datetime
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Cell 5: LearnMate Agent System
# ================================
# 4. LEARNMATE AGENT SYSTEM
# ================================

@dataclass
class LearningPath:
    user_id: int
    domain: str
    current_level: str
    goal: str
    courses: List[Dict]
    progress: float = 0.0
    estimated_duration: int = 0
    next_milestone: str = ""

class LearnMateAgent:
    """
    Main Agentic AI system for personalized learning pathways
    Integrates with IBM Watsonx AI Studio and Granite models
    """

    def __init__(self, users_df: pd.DataFrame, courses_df: pd.DataFrame, ai_model: IBMGraniteAI):
        self.users = users_df
        self.courses = courses_df
        self.ai_model = ai_model
        self.learning_paths = {}
        self.interaction_history = []

    def interact_with_user(self, user_input: str, user_id: int = None) -> str:
        """Handle natural language interaction with users"""
        timestamp = datetime.now().strftime("%Y-%m-%d %H:%M:%S")

        # Analyze user intent using IBM Granite
        intent = self.ai_model.understand_user_intent(user_input)

        # Generate contextual response
        response = self.ai_model.generate_response(user_input, context=intent)

        # Log interaction
        self.interaction_history.append({
            'timestamp': timestamp,
            'user_id': user_id,
            'input': user_input,
            'intent': intent,
            'response': response
        })

        return response

    def assess_user_profile(self, user_id: int) -> Dict:
        """Analyze user profile and learning preferences"""
        user = self.users[self.users['UserID'] == user_id].iloc[0]

        profile = {
            'name': user['Name'],
            'primary_interest': user['Interest'],
            'skill_level': user['SkillLevel'],
            'goal': user['Goal'],
            'preferred_duration': user['PreferredDuration'],
            'learning_style': user['LearningStyle'],
            'experience': user['PreviousExperience']
        }

        return profile

    def filter_courses_by_criteria(self, domain: str, skill_level: str,
                                 max_duration: int = None) -> pd.DataFrame:
        """Filter courses based on user criteria"""
        filtered = self.courses[self.courses['Domain'] == domain].copy()

        # Progressive skill level filtering
        if skill_level == 'Beginner':
            filtered = filtered[filtered['Level'].isin(['Beginner'])]
        elif skill_level == 'Intermediate':
            filtered = filtered[filtered['Level'].isin(['Beginner', 'Intermediate'])]
        else:  # Advanced
            filtered = filtered[filtered['Level'].isin(['Intermediate', 'Advanced'])]

        # Duration filtering if specified
        if max_duration:
            filtered = filtered[filtered['Duration(hrs)'] <= max_duration]

        # Sort by rating and level appropriateness
        filtered = filtered.sort_values(['Level', 'Rating'], ascending=[True, False])

        return filtered

    def create_learning_pathway(self, user_id: int, max_courses: int = 6) -> LearningPath:
        """Generate personalized learning pathway"""
        profile = self.assess_user_profile(user_id)

        # Get suitable courses
        suitable_courses = self.filter_courses_by_criteria(
            domain=profile['primary_interest'],
            skill_level=profile['skill_level']
        )

        # Create progressive learning sequence
        pathway_courses = []
        levels = ['Beginner', 'Intermediate', 'Advanced']

        for level in levels:
            level_courses = suitable_courses[suitable_courses['Level'] == level]
            if not level_courses.empty:
                # Select best courses for this level
                selected = level_courses.head(2).to_dict('records')
                pathway_courses.extend(selected)

                if len(pathway_courses) >= max_courses:
                    break

        # Calculate pathway metrics
        total_duration = sum(course['Duration(hrs)'] for course in pathway_courses)
        next_milestone = pathway_courses[0]['Title'] if pathway_courses else "No courses found"

        # Create learning path object
        learning_path = LearningPath(
            user_id=user_id,
            domain=profile['primary_interest'],
            current_level=profile['skill_level'],
            goal=profile['goal'],
            courses=pathway_courses,
            estimated_duration=total_duration,
            next_milestone=next_milestone
        )

        self.learning_paths[user_id] = learning_path
        return learning_path

    def simulate_progress_update(self, user_id: int, completed_course_id: int) -> LearningPath:
        """Simulate user completing a course and update pathway"""
        if user_id not in self.learning_paths:
            return None

        path = self.learning_paths[user_id]

        # Find and mark course as completed
        for i, course in enumerate(path.courses):
            if course['CourseID'] == completed_course_id:
                course['completed'] = True
                course['completion_date'] = datetime.now().strftime("%Y-%m-%d")
                break

        # Update progress
        completed_courses = sum(1 for course in path.courses if course.get('completed', False))
        path.progress = (completed_courses / len(path.courses)) * 100

        # Update next milestone
        next_course = next((course for course in path.courses if not course.get('completed', False)), None)
        path.next_milestone = next_course['Title'] if next_course else "Pathway Complete! 🎉"

        # Adaptive recommendation: suggest new courses based on progress
        if completed_courses >= len(path.courses) // 2:  # Halfway through
            self._add_advanced_courses(path)

        return path

    def _add_advanced_courses(self, path: LearningPath):
        """Add more advanced courses based on progress"""
        advanced_courses = self.filter_courses_by_criteria(
            domain=path.domain,
            skill_level='Advanced'
        ).head(2).to_dict('records')

        # Add new courses to pathway
        for course in advanced_courses:
            if not any(c['CourseID'] == course['CourseID'] for c in path.courses):
                path.courses.append(course)

        # Recalculate duration
        path.estimated_duration = sum(course['Duration(hrs)'] for course in path.courses)

    def generate_pathway_visualization(self, user_id: int) -> None:
        """Create visual representation of learning pathway"""
        if user_id not in self.learning_paths:
            print("No learning path found for this user.")
            return

        path = self.learning_paths[user_id]

        # Create pathway visualization
        fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(15, 12))
        fig.suptitle(f'LearnMate AI - Learning Dashboard for {path.domain}', fontsize=16, fontweight='bold')

        # 1. Course Progression Timeline
        course_names = [course['Title'][:20] + '...' if len(course['Title']) > 20 else course['Title']
                       for course in path.courses]
        course_levels = [course['Level'] for course in path.courses]
        course_durations = [course['Duration(hrs)'] for course in path.courses]
        completed = [course.get('completed', False) for course in path.courses]

        colors = ['green' if comp else 'lightblue' for comp in completed]
        ax1.barh(course_names, course_durations, color=colors)
        ax1.set_xlabel('Duration (hours)')
        ax1.set_title('Learning Pathway Timeline')
        ax1.grid(axis='x', alpha=0.3)

        # 2. Progress Pie Chart
        completed_hours = sum(course['Duration(hrs)'] for course in path.courses
                            if course.get('completed', False))
        remaining_hours = path.estimated_duration - completed_hours

        ax2.pie([completed_hours, remaining_hours],
                labels=['Completed', 'Remaining'],
                colors=['#2ecc71', '#ecf0f1'],
                autopct='%1.1f%%',
                startangle=90)
        ax2.set_title(f'Progress Overview\n{path.progress:.1f}% Complete')

        # 3. Skill Level Distribution
        level_counts = pd.Series(course_levels).value_counts()
        ax3.bar(level_counts.index, level_counts.values, color=['#3498db', '#e74c3c', '#f39c12'])
        ax3.set_title('Course Difficulty Distribution')
        ax3.set_ylabel('Number of Courses')

        # 4. Learning Statistics
        stats_data = {
            'Total Courses': len(path.courses),
            'Completed': sum(completed),
            'Remaining': len(path.courses) - sum(completed),
            'Total Hours': path.estimated_duration,
            'Completed Hours': completed_hours
        }

        ax4.axis('off')
        stats_text = '\n'.join([f'{key}: {value}' for key, value in stats_data.items()])
        ax4.text(0.1, 0.8, stats_text, fontsize=12, verticalalignment='top',
                bbox=dict(boxstyle='round', facecolor='lightgray', alpha=0.8))
        ax4.text(0.1, 0.4, f'Next Milestone:\n{path.next_milestone}', fontsize=10,
                bbox=dict(boxstyle='round', facecolor='lightblue', alpha=0.8))
        ax4.set_title('Learning Statistics')

        plt.tight_layout()
        plt.show()

# Initialize LearnMate Agent
print("\n🧠 Initializing LearnMate Agent...")
learnmate = LearnMateAgent(user_profiles, course_data, granite_ai)
print("✅ LearnMate Agent Ready!")
```

```python
# Cell 10: Feedback Loop Simulation
# ================================
# 9. FEEDBACK LOOP SIMULATION
# ================================

print("\n" + "="*60)
print("🔄 AI FEEDBACK LOOP & CONTINUOUS IMPROVEMENT")
print("="*60)

def simulate_feedback_collection(user_id: int):
    """Simulate collecting user feedback and improving recommendations"""

    if user_id not in learnmate.learning_paths:
        print("No learning path found for feedback simulation.")
        return

    path = learnmate.learning_paths[user_id]

    # Simulate user feedback
    feedback_scenarios = [
        {
            'type': 'course_difficulty',
            'feedback': 'too_easy',
            'course_id': path.courses[0]['CourseID'],
            'action': 'recommend_advanced'
        },
        {
            'type': 'learning_pace',
            'feedback': 'too_fast',
            'adjustment': 'extend_timeline'
        },
        {
            'type': 'content_preference',
            'feedback': 'more_projects',
            'action': 'prioritize_project_based'
        }
    ]

    print(f"📝 Processing feedback for user {user_id}...")

    for feedback in feedback_scenarios:
        print(f"\n🔹 Feedback Type: {feedback['type']}")
        print(f"   User Input: {feedback['feedback']}")

        if feedback['type'] == 'course_difficulty' and feedback['feedback'] == 'too_easy':
            # AI adapts by recommending more challenging courses
            advanced_courses = course_data[
                (course_data['Domain'] == path.domain) &
                (course_data['Level'] == 'Advanced')
            ].head(2)

            print(f"   🤖 AI Adaptation: Adding advanced courses")
            for _, course in advanced_courses.iterrows():
                print(f"      + {course['Title']} ({course['Level']})")

        elif feedback['type'] == 'learning_pace':
            print(f"   🤖 AI Adaptation: Adjusting learning timeline")
            print(f"      Extending recommended study time per course")

        elif feedback['type'] == 'content_preference':
            print(f"   🤖 AI Adaptation: Prioritizing project-based courses")
            project_courses = course_data[
                (course_data['Domain'] == path.domain) &
                (course_data['ProjectIncluded'].notna())
            ]
            print(f"      Found {len(project_courses)} project-based courses")

# Simulate feedback for our demo user
simulate_feedback_collection(1)
```

```python
# Cell 11: Real-time Recommendation Engine
# ================================
# 10. REAL-TIME RECOMMENDATION ENGINE
# ================================

print("\n" + "="*60)
print("⚡ REAL-TIME RECOMMENDATION ENGINE")
print("="*60)

class RealtimeRecommendationEngine:
    """Real-time course recommendation system"""

    def __init__(self, learnmate_agent):
        self.agent = learnmate_agent
        self.trending_courses = self._get_trending_courses()
        self.peer_recommendations = self._generate_peer_recommendations()

    def _get_trending_courses(self):
        """Get trending courses based on ratings and enrollment simulation"""
        # Simulate trending based on high ratings and recent popularity
        trending = self.agent.courses[
            self.agent.courses['Rating'] >= 4.5
        ].sample(n=min(10, len(self.agent.courses)))
        return trending

    def _generate_peer_recommendations(self):
        """Generate peer-based recommendations"""
        peer_recs = {}
        for domain in self.agent.courses['Domain'].unique():
            top_courses = self.agent.courses[
                self.agent.courses['Domain'] == domain
            ].nlargest(3, 'Rating')
            peer_recs[domain] = top_courses['Title'].tolist()
        return peer_recs

    def get_instant_recommendations(self, user_id: int, context: str = "general"):
        """Get instant recommendations based on context"""
        if user_id not in self.agent.learning_paths:
            return []

        path = self.agent.learning_paths[user_id]

        if context == "trending":
            trending_in_domain = self.trending_courses[
                self.trending_courses['Domain'] == path.domain
            ]
            return trending_in_domain.head(3).to_dict('records')

        elif context == "peer_suggestions":
            return self.peer_recommendations.get(path.domain, [])

        elif context == "skill_gap":
            # Identify skill gaps and recommend accordingly
            completed_skills = set()
            for course in path.courses:
                if course.get('completed', False):
                    if 'Skills' in course:
                        completed_skills.update(course['Skills'].split('/'))

            # Find courses that teach new skills
            skill_gap_courses = []
            for _, course in self.agent.courses.iterrows():
                if course['Domain'] == path.domain:
                    course_skills = set(course.get('Skills', '').split('/'))
                    if course_skills - completed_skills:  # Has new skills
                        skill_gap_courses.append(course.to_dict())

            return skill_gap_courses[:3]

        return []

# Initialize real-time engine
rt_engine = RealtimeRecommendationEngine(learnmate)

print("\n🔥 Trending Courses:")
trending = rt_engine.get_instant_recommendations(1, "trending")
for course in trending:
    print(f"   📈 {course['Title']} - Rating: {course['Rating']}")

print("\n👥 Peer Recommendations for Frontend Development:")
peer_recs = rt_engine.get_instant_recommendations(1, "peer_suggestions")
for i, rec in enumerate(peer_recs, 1):
    print(f"   {i}. {rec}")

print("\n🎯 Skill Gap Analysis:")
skill_gaps = rt_engine.get_instant_recommendations(1, "skill_gap")
for course in skill_gaps:
    print(f"   🔧 {course['Title']} - New Skills: {course.get('Skills', 'N/A')}")
```
